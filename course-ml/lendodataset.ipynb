```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(dataset[:50, 0], c='Red', ls=':', marker='s', ms=8)
```

```python
dataset[50:100,0] #dados primeira coluna iris-versicolor
```

```python
plt.plot(dataset[:50, 0], c='Red', ls=':', marker='s', ms=6, label='Comp. Spala Iris-Setosa')
plt.plot(dataset[50:100, 0], c='Black', ls=':', marker='o', ms=6, label='Comp. Spala Iris-Versicolor')
plt.plot(dataset[100:150, 0], c='Green', ls=':', marker='^', ms=6, label='Comp. Spala Iris-Virginica')
plt.legend()
plt.show()
```
